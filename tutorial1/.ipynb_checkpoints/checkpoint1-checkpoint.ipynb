```python
import numpy
import pandas
pandas.options.mode.chained_assignment = None
import matplotlib.pyplot as mpl
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
education = pandas.read_csv('../data/states_edu.csv')
education["PRIMARY_KEY"].isna()
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
education.dropna(subset=['AVG_MATH_4_SCORE'], inplace=True)
math = education.loc[::][['PRIMARY_KEY', 'STATE', 'YEAR', 'ENROLL', 'GRADES_PK_G', 'GRADES_KG_G',
       'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G',
       'GRADES_9_12_G', 'GRADES_ALL_G', 'AVG_MATH_4_SCORE']]
michigan_math = math.loc[math["STATE"] == "MICHIGAN"][["STATE", "AVG_MATH_4_SCORE"]]
ohio_math = math.loc[math["STATE"] == "OHIO"][["STATE", "AVG_MATH_4_SCORE"]]
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = education[["GRADES_4_G", "AVG_MATH_8_SCORE", "TOTAL_EXPENDITURE_PER_STUDENT"]].dropna()
y = education.loc[X.index]["AVG_MATH_4_SCORE"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.score(X_test, y_test)) # r_squared
print(numpy.mean(model.predict(X_test) - y_test)) # mean
print(numpy.mean((model.predict(X_test) - y_test)**2)**0.5) # root mean squared error
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = "TOTAL_EXPENDITURE_PER_STUDENT"
f = mpl.figure(figsize = (12, 6))
mpl.scatter(X_train[col_name], y_train, color = "red")
mpl.scatter(X_train[col_name], model.predict(X_train), color = "green")
mpl.legend(["True Training", "Predicted Training"])
mpl.xlabel("Total Spending Per Student")
mpl.ylabel("Average 4th Grade Math Score")
mpl.title("Model Behavior On Training Set")
```

```python
col_name = "TOTAL_EXPENDITURE_PER_STUDENT"
f = mpl.figure(figsize = (12, 6))
mpl.scatter(X_test[col_name], y_test, color = "blue")
mpl.scatter(X_test[col_name], model.predict(X_test), color = "purple")
mpl.legend(["True testing", "Predicted testing"])
mpl.xlabel("Total Spending Per Student")
mpl.ylabel("Average 4th Grade Math Score")
mpl.title("Model Behavior on Testing Set")
```

<h2> Summary </h2>

The model has a r_squared of 0.8574, indicating that there is a significant correlation between total spending per student and the average 4th grade math score. However, the root mean squared error of our model is considered unsatistfactory under typical data science constraints, as the value of 4.0420 is higher than an ideal value of less than 0.75. Despite this, it can still reasonably be concluded the average math score for 4th graders and total spending per student has a moderately strong correlation.
